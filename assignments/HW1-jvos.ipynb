## Exercise 5

```julia
function normvec(v, normtype = 2)
    function norm2(v)
        return sqrt(sum(v.^ 2))
    end
    function norm1(v)
        return sum(abs.(v))
    end
    function norminf(v)
        return maximum(abs.(v))
    end
    
    normtype==1 && return norm1(v)
    normtype==2 && return norm2(v)
    normtype==Inf && return norminf(v)
                
end
    
```

```julia
# Example
v = [1;2;3]
```

```julia
abs(normvec(v, 2) - norm(v, 2))
```

```julia
abs(normvec(v, 1) - norm(v, 1))
```

```julia
abs(normvec(v, Inf) - norm(v, Inf))
```

## Exercise 21

```julia
function normmat(A, normtype = 1)
    function normmat1(A)
        norm = 0.0
        for j = 1:size(A, 2)
            col_sum = sum(abs.(A[:, j]))
            if(norm < col_sum)
                norm = col_sum
            end
        end
        return norm
    end
    function normmatinf(A)
        norm = 0.0
        for j = 1:size(A, 1)
            row_sum = sum(abs.(A[j, :]))
            if(norm < row_sum)
                norm = row_sum
            end
        end
        return norm
    end
    
    normtype==1 && return normmat1(A)
    normtype==Inf && return normmatinf(A)
                
end
    
```

```julia
# Example
A = rand(3,4)
```

```julia
abs(opnorm(A, 1) - normmat(A, 1))
```

```julia
abs(opnorm(A, Inf) - normmat(A, Inf))
```

## Exercise 7

```julia
function plot_unit_ball(p)
    θ = LinRange(0, 2π, 150)
    x = cos.(θ)
    y = sin.(θ)

    norm_vals = (abs.(x).^p + abs.(y).^p).^(1/p)
    x_normalized = x ./ norm_vals
    y_normalized = y ./ norm_vals

    trace = scatter(x=x_normalized, y=y_normalized, mode="lines", name="p = $p-norm")
    return trace
end

# Example: Plot unit balls for different p-norms
traces = [plot_unit_ball(1.1), plot_unit_ball(3.5), plot_unit_ball(3), plot_unit_ball(10)]

layout = Layout(
    title="Unit Balls for Different p-norms",
    xaxis_title="x1",
    yaxis_title="x2",
    legend=attr(x=1.02, y=0.95),
    width = 800,
    height = 600,
    xaxis_range=[-1.2, 1.2],
    yaxis_range=[-1.2, 1.2]
)

plot(traces, layout)
```

#### As $n\to 1$, the figure is approaching a diamond, while as $n\to \infty$, the figure is approaching a square of sides of size 1.
